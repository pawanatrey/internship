```python
#Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and 
# salary filter.
#You have to scrape data for ‚ÄúData Scientist‚Äù designation for first 10 job results.
#You have to scrape the job-title, job-location, company name, experience required.
#The location filter to be used is ‚ÄúDelhi/NCR‚Äù. The salary filter to be used is ‚Äú3-6‚Äù lakhs
#The task will be done as shown in the below steps:
#1. first get the web page https://www.naukri.com/
#2. Enter ‚ÄúData Scientist‚Äù in ‚ÄúSkill, Designations, and Companies‚Äù field.
#3. Then click the search button.
#4. Then apply the location filter and salary filter by checking the respective boxes
#5. Then scrape the data for the first 10 jobs results you get.
#6. Finally create a dataframe of the scraped data.
```

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By 
import time 
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.naukri.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Analyst')
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
location=driver.find_element(By.XPATH,"/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[4]/div[2]/div[2]/label/p/span[1]")
location.click()
```

```python
salary=driver.find_element(By.XPATH,"/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[5]/div[2]/div[2]/label/p/span[1]")
salary.click()
```

```python
title=[]
desig=driver.find_elements(By.XPATH,'//a[@class="title "]')
for i in desig[0:10]:
    title.append(i.text)
title    
```

```python
loca=[]
place=driver.find_elements(By.XPATH,'//span[@class="locWdth"]')
for i in place[0:10]:
    loca.append(i.text)
loca
```

```python
salaryy=[]
amt=driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-rupee sal"]')
for i in amt[0:10]:
    salaryy.append(i.text)
salaryy
```

```python
df=pd.DataFrame({'Title':title,'Location':loca,'Salary':salaryy})
df
```
