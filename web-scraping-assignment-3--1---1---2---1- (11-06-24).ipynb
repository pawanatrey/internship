```python
# Q2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results 
# and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the 
# products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", 
# "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing
# for any of the product then replace it by
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.amazon.in/")
```

```python
search_item=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
search_item.send_keys("guitar")
```

```python
search_clk=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
search_clk.click()
```

```python
guitar_clk=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div[2]/div/div[3]/span/div[1]/div/div/div[4]/ul/span/span[3]/li/span/a/span")
guitar_clk.click()
```

```python
search=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div[2]/div/div[3]/span/div[1]/div/div/div[4]/ul/span/span[3]/li/span/a/span")
search.click()
```

```python
# Extract all the url's from first 3 pages
import time
product_url=[]
start=0
end=3
for i in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
for j in url:
    product_url.append(j.get_attribute('href'))
search1=driver.find_element(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
search1.click()
time.sleep(2)
product_url
```

```python
# Now click on product url and extract all the information.
brand=[]
Product_name=[]
price=[]
available=[]   
for link in product_url: 
    driver.get(link)
    time.sleep(2)
    
    try:
        bra=driver.find_element(By.XPATH,'//*[@id="productTitle"]')
        brand_name.append(bra.text)  
    
    except NoSuchElementException:
        brand.append('-')            
    
    try:
        name=driver.find_element(By.XPATH,'//*[@id="productTitle"]')
        Product_name.append(name.text)  
    
    except NoSuchElementException:
        Product_name.append('-')            

    try:
        pri=driver.find_element(By.XPATH,'//*[@id="productTitle"]')
        price.append(pri.text)  
    
    except NoSuchElementException:
        price.append('-')            
    
    try:
        avail=driver.find_element(By.XPATH,'//*[@id="productTitle"]')
        available.append(avail.text)  
    
    except NoSuchElementException:
        available.append('-')              
```

```python
df=pd.DataFrame({'Brand':brand,'Product_Name':Product_name,'Product_Price':price,'Product_Availability':available,
                 'Product_URL':product_url})
df.head(10)
```

```python
df.to_csv('guitar_copy.csv') # file saved in CSV file 
```
