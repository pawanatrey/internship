```python
# Q4- Write a python program to scrape first 10 product details which include product name , price , Image URL from
# https://www.bewakoof.com/bestseller?sort=popular .
```

```python
data=requests.get("https://www.bewakoof.com/bestseller?sort=popular")
data
```

```python
soup=BeautifulSoup(data.content)
soup
```

```python
Product_name=[]
for i in soup.find_all('h3',class_="brand-name undefined"):
    Product_name.append(i.text)
Product_name
```

```python
Product_Price=[]
for i in soup.find_all('div',class_="discountedPriceText clr-p-black false"):
    Product_Price.append(i.text)
Product_Price
```

```python
all_images=[]

for i in soup.find_all("img",class_="productImgTag"):
    all_images.append(i['src'])
all_images
```

```python
df=pd.DataFrame({'Product Name':Product_name,'Product Price':Product_Price,'Image URL':all_images})
df.head(10)
```
